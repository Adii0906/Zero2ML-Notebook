```python
import pandas as  pd
df=pd.read_csv('data.csv')
df.head(5)
```

```python
df.sort_values(by='Actor',ascending=False).head(5) #It will sort the index one by one 
```

```python
# Convert data type
df["Age"] = df["Age"].astype(int)
```

```python
df2=df.sort_values(['Actor','IMDb']).copy()
df2.head(5)
```

```python
df2.reset_index(drop=True,inplace=True) #It reset the index from 0,1,2,3,4/....
df2.head(5)
```

```python
df2.sort_index().head(5) 
```

```python
df2['Rank']=df2['IMDb'].rank(ascending=False,method='dense')
df2.head(5)
```
