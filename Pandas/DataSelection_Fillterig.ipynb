```python
import pandas as pd
df=pd.read_csv('data.csv')
df.head()
```

```python
# Column Selection and Filtering in Pandas
df['Actor'].head(5)
```

```python
# Suppose u want actors and imdb scores
# this is just a view of the data not a new dataframe
# note: this dosent change the original dataframe
df[['Actor','IMDb']].head(5)
```

```python
# iloc is used for selecting rows and columns by position
df.iloc[1]
```

```python
df.loc[8,'IMDb']  # accessing a particular value using label
```

```python
df.loc[0:2,['Actor','IMDb']]  # accessing multiple rows and columns using label
```

```python
df.at[0,'Actor']  # faster access for a single value
# Reason : we can use iloc and loc for single value access but at and iat are faster for one element access
df.iat[1,2]  # faster access for a single value using position
```

```python
# Filtering DataFrames
# Run the cells below to see various filtering operations
df[df['IMDb']>8]  # all rows where imdb is greater than 8
df[df['IMDb']>7]['Actor']  # actors with imdb greater than 7
df[df['IMDb']>7][['Actor','IMDb']]  # actors with imdb greater than 7 with their imdb scores
df[df["IMDb"]>6 & (df['Year']>2000)].head(5)  # multiple conditions using & (and) and | (or)
```

```python
# query is easy and readable and efficient for complex queries
# sql like syntax
# do not use logical operators use and , or instead use == != < > <= >=
df.query('IMDb > 8')  # using query method for filtering
df.query("IMDb >7 & Year > 2000").head(4)  # multiple conditions using query method
df.query
```
