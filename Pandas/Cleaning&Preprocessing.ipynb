Data Cleaning and PreProcessing
Suppose you have  many data and that data are messy 
Pandas Gives you a power to clean and transform the data before analysis it 


```python
import pandas as pd
df=pd.read_csv('data_cleaning_example.csv')
df
```

```python
df.isnull().sum() # to check how many null values are there in each column
```

```python
df.dropna(axis=1) #drops columns with any null values
# df.dropna(axis=0) #drops rows with any null values
df.fillna(0) # fills all null values with 0
df['Age'].fillna(df['Age'].mean()) # fills null values in 'Age' column with mean of that column
df.fillna(method='ffill') # forward can be used with axis=0 or axis=1
df.fillna(method='bfill') # backward can be used with axis=0 or axis=1
```

```python
# Removing the dupliacate values
df.duplicated() # to check how many duplicate rows are there
#Flase means unique row and True means duplicate row
```

```python
# suppose u dont want duplicate rows
df.drop_duplicates() # this will remove duplicate rows and keep only unique rows
print("Duplicates Droped",df) # original dataframe remains unchanged
```

```python
df.duplicated(subset=['Name','Age']) # to check duplicates based on specific columns
```

```python
df['City'].str.contains('New')  # to filter rows where 'City' column contains the substring 'New'
```
