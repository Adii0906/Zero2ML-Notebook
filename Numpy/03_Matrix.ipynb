```python
import numpy as np

# Matrix multiplication
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])
C = A @ B
print("Matrix Multiplication:\n", C)

# Transpose
print("Transpose of A:\n", A.T)

# -------- Real world ML Normalization Example --------
# Dataset: [Hours studied, Exam score]
data = np.array([[2, 50], [5, 65], [7, 80], [10, 95]])
X = data[:, 0]  # hours
y = data[:, 1]  # score

# Min-Max normalization
X_norm = (X - X.min()) / (X.max() - X.min())
print("Normalized X:", X_norm)

# Dot product example (weighted score)
weights = np.array([0.7, 0.3])
student = np.array([85, 70])     # exam + attendance
weighted_score = np.dot(weights, student)
print("Weighted Score:", weighted_score)
```
