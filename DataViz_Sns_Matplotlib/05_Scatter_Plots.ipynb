```python
import matplotlib.pyplot as plt
cgpa = [3.5, 3.6, 3.7, 3.8, 3.9, 4.0]
subject_scores = [70, 75, 80, 85, 90, 95]
plt.scatter(cgpa, subject_scores, color='blue', marker='o')
plt.title('CGPA vs Subject Scores')
plt.xlabel('CGPA')
plt.ylabel('Subject Scores')
plt.show()
```

```python
x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

# Create scatter plot
plt.scatter(x, y, color='blue', marker='o', s=100, alpha=0.7)

# Add labels and title
plt.xlabel("X-axis values")
plt.ylabel("Y-axis values")
plt.title("Simple Scatter Plot")

# Show plot
plt.show()
```

```python
section1_scores = [88, 92, 95, 70, 65, 80, 85, 90]
section2_scores = [75, 85, 80, 60, 70, 75, 80, 95]
section1_students = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
section2_students = ['I', 'J', 'K', 'L', 'M', 'N', 'O', 'P']

plt.scatter(section1_students, section1_scores, color='blue', marker='o', label='Section 1')
plt.scatter(section2_students, section2_scores, color='green', marker='^', label='Section 2')
plt.title('Student Scores in Two Sections')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.legend()
plt.show()
```
