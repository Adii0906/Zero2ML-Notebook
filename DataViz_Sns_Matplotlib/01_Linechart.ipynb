```python
import matplotlib.pyplot as plt # Importing the matplotlib library for plotting
# if module not installed, run: pip install matplotlib
```

```python
# Suppose u want to plot a simple line chart using matplotlib. Here's how you can do it:
x = [1, 2, 3, 4, 5]   # X-axis data
y = [2, 3, 5, 7, 11]   # Y-axis data
plt.plot(x, y)   # Plotting the line chart
plt.title("Simple Line Chart")   # Adding a title
plt.xlabel("X-axis")   # Labeling the X-axis
plt.ylabel("Y-axis")   # Labeling the Y-axis
plt.show()   # Displaying the plot
```

```python
# suppose we have student marks and we want to plot a line chart to show the trend of marks over different subjects. Here's how you can do it:
subjects = ['Math', 'Science', 'English', 'History', 'Art']  # Subjects
marks = [85, 90, 78, 88, 92]  # Corresponding marks
plt.plot(subjects, marks, marker='o')  # Plotting the line chart with markers
# marker is used to highlight each data point on the line
plt.title("Student Marks Across Subjects")  # Adding a title
plt.xlabel("Subjects")  # Labeling the X-axis
plt.ylabel("Marks")  # Labeling the Y-axis
```

```python
# suppose we have 2 student marks
John_marks = [85, 90, 78, 88, 92]
lily_marks = [80, 85, 82, 90, 95]
plt.tight_layout()  # Adjust layout to prevent overlap
plt.plot(subjects, John_marks, marker='o', label='John')  # Plotting John's marks
plt.plot(subjects, lily_marks, marker='s', label='Lily')  # Plotting Lily's marks
```
