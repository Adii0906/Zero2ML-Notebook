```python
import matplotlib.pyplot as plt

cgpa = [8.5, 9.0, 7.8, 8.2, 9.1]
semesters = ['Sem1', 'Sem2', 'Sem3', 'Sem4', 'Sem5']

plt.title("Student CGPA over Semesters")
plt.pie(cgpa, labels=semesters)
plt.show()
```

```python
# 3d pie chart
plt.pie(
    cgpa,
    labels=semesters,
    explode=[0.1, 0, 0, 0, 0],
    shadow=True,
)
```

```python
plt.pie(
    cgpa,
    labels=semesters,
    explode=[0.1, 0, 0, 0, 0],
    autopct='%1.1f%%', # Show percentage
    shadow=True,
)
```
