```python
import matplotlib.pyplot as plt
studies = ['Math', 'Physics', 'Chemistry', 'Biology', 'History']
scores = [88, 92, 85, 90, 78]
plt.bar(studies, scores)
plt.xlabel('Subjects')
plt.ylabel('Scores')
plt.title('Student Scores by Subject')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

student_names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
student_scores = [95, 85, 78, 88, 92]

x = np.arange(len(student_names))  # positions for bars
width = 0.35

plt.bar(x, student_scores, width, label='Scores', color='skyblue')
plt.xlabel('Students')
plt.ylabel('Scores')
plt.xticks(x, student_names)
plt.title('Student Scores by Name')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

student_names = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
student_scores = [95, 85, 78, 88, 92]
subjects = ['Math', 'Science', 'History', 'Art', 'Physical Ed']

# Create bar chart
plt.bar(student_names, student_scores, color='lightcoral')

# Add annotations above each bar
for i in range(len(student_names)):
    plt.text(i, student_scores[i] + 2, f"{student_scores[i]} ({subjects[i]})", ha='center')

plt.xlabel('Students')
plt.ylabel('Scores')
plt.title('Student Scores with Annotations')
plt.show()
```
